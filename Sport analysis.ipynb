```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_excel(r'\Users\Maddy\Downloads\Oly.xlsx')
```

# Q1 Which countries have won the most medals?


```python
q1=df.groupby('Country').sum('Total Medals').reset_index()
print(q1.loc[q1['Total Medals'].idxmax(),["Country","Total Medals"]])
sns.barplot(data=q1,x='Country',y='Total Medals') ### here you can't see the graph clear so i applied some more code to show 
## graph in a very clear view....
```

## so USA won 1312 Total medals 

```python
Q1=q1.nlargest(n=8, columns='Total Medals')
print(Q1.loc[Q1['Total Medals'].idxmax(),["Country","Total Medals"]])
sns.barplot(data=Q1,x='Country',y='Total Medals')

```

# Q2 Which countries have won the most gold medals?

```python
q2=q1.loc[q1['Gold Medals'].idxmax(),['Country','Gold Medals']]
print(q2)
Q2=q1.nlargest(n=8,columns='Gold Medals')
sns.barplot(data=Q2,x='Country',y='Gold Medals')
```

## Hence again max Gold medal won by USA
